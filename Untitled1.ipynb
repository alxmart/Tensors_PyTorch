## ** Tensors - PyTorch

```python
import torch
import numpy as np

lista = [[1,2,3],
         [4,5,6]]

tns = torch.Tensor(lista)
print(tns.dtype)
print(f"{tns} \n")

tnsf = torch.FloatTensor(lista)
print(tnsf.dtype)
print(f"{tnsf} \n")

tnsd = torch.DoubleTensor(lista)
print(tnsd.dtype)
print(f"{tnsd} \n")

tnsd = torch.LongTensor(lista)
print(tnsd.dtype)
print(f"{tnsd} \n")
```

```python
# A partir de arrays NumPy

arr = np.random.rand(3,4)
arr = arr.astype(int)
tns2 = torch.from_numpy(arr)

print(arr)
print(f"{tns2} \n")

print(arr.dtype)
print(tns2.dtype)
```

```python
# Tensores inicializados

tns0 = torch.zeros(4,5)
print(f"{tns0} \n")

tns1 = torch.ones(2,3)
print(f"{tns1} \n")

tnsr = torch.rand(3,3)
print(f"{tnsr} \n")




```

```python
# Tensor para Array NumPy
print(type(tnsr))
arrnp = tnsr.data.numpy()
print(arrnp)
print(type(arrnp))
```

```python
# Indexação
print(f"{tnsr}\n")

tnsr[0, 2] = -10

print(f"{tnsr}\n")

print(f"{tnsr[0:2]}\n") # linhas 0 e 1

print(f"{tnsr[:, 2]}\n") # linhas da coluna 2

print(f"{tnsr[0,2]}\n")

print(f"{tnsr[0,2].size()}\n")

```

```python
# Operações com tensores

tns_fatiado = tnsr[0:2, :]

print(tns_fatiado.shape)

print(tns1.shape)
print("")

print(f"{tns_fatiado}\n")
print(f"{tns1}\n")


tns_soma = tns1 + tns_fatiado
print(f"{tns_soma}\n")


tns_sub = tns1 - tns_fatiado
print(f"{tns_sub}\n")


tns_mult = tns1 * tns_fatiado
print(f"{tns_mult}\n")


tns_div = tns1 / tns_fatiado
print(f"{tns_div}\n")


# Multiplicação de matrizes

tns_mm = torch.mm(tns1, tns_fatiado.T)
print(f"{tns_mm}\n")


tns_mm2 = tns1.mm(tns_fatiado.T)
print(f"{tns_mm2}\n")



```

```python
# Função .size() e .view()

print(tns_fatiado.shape)

print(tns_fatiado.size())
```

```python
tnsx = torch.randn(2, 2, 3)
print(tnsx)
```

```python
tnsv = tnsx.view(12) # Faz um "achatamento"
print(tnsv)
```

```python
tnsv = tnsx.view(-1) # Faz "achatamento"
print(tnsv)
```

```python
tnsv2 = tnsx.view(4, 3)
print(tnsv2)
```

```python
tnsv3 = tnsx.view(tnsx.size(0),-1) # Mantem a primeira dimensão (2 linhas) e "achata" o resto
print(tnsv3)
```
