## ** Tensors - PyTorch

```python
import torch
import numpy as np

lista = [[1,2,3],
         [4,5,6]]

tns = torch.Tensor(lista)
print(tns.dtype)
print(f"{tns} \n")

tnsf = torch.FloatTensor(lista)
print(tnsf.dtype)
print(f"{tnsf} \n")

tnsd = torch.DoubleTensor(lista)
print(tnsd.dtype)
print(f"{tnsd} \n")

tnsd = torch.LongTensor(lista)
print(tnsd.dtype)
print(f"{tnsd} \n")
```

```python
# A partir de arrays NumPy

arr = np.random.rand(3,4)
arr = arr.astype(int)
tns2 = torch.from_numpy(arr)

print(arr)
print(f"{tns2} \n")

print(arr.dtype)
print(tns2.dtype)
```
