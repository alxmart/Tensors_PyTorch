## ** Tensors - PyTorch

```python
import torch
import numpy as np

lista = [[1,2,3],
         [4,5,6]]

tns = torch.Tensor(lista)
print(tns.dtype)
print(f"{tns} \n")

tnsf = torch.FloatTensor(lista)
print(tnsf.dtype)
print(f"{tnsf} \n")

tnsd = torch.DoubleTensor(lista)
print(tnsd.dtype)
print(f"{tnsd} \n")

tnsd = torch.LongTensor(lista)
print(tnsd.dtype)
print(f"{tnsd} \n")
```

```python
# A partir de arrays NumPy

arr = np.random.rand(3,4)
arr = arr.astype(int)
tns2 = torch.from_numpy(arr)

print(arr)
print(f"{tns2} \n")

print(arr.dtype)
print(tns2.dtype)
```

```python
# Operações com tensores

tns_fatiado = tnsr[0:2, :]

print(tns_fatiado.shape)

print(tns1.shape)
print("")

print(f"{tns_fatiado}\n")
print(f"{tns1}\n")


tns_soma = tns1 + tns_fatiado
print(f"{tns_soma}\n")


tns_sub = tns1 - tns_fatiado
print(f"{tns_sub}\n")


tns_mult = tns1 * tns_fatiado
print(f"{tns_mult}\n")


tns_div = tns1 / tns_fatiado
print(f"{tns_div}\n")


# Multiplicação de matrizes

tns_mm = torch.mm(tns1, tns_fatiado.T)
print(f"{tns_mm}\n")


tns_mm2 = tns1.mm(tns_fatiado.T)
print(f"{tns_mm2}\n")



```

```python
import torch

tns = torch.randn(9, 12)
print(tns)
print("")

tns1 = tns[0:5, 0:4]
# 5: comece na linha de índice 5 e pegue todas as linhas até o final do tensor.
# 4: comece na coluna de índice 4 e pegue todas as colunas até o final do tensor.
# 4 linhas e 8 colunas
tns2 = tns[5:, 4:]
print(f"Tensor tns1 \n {tns1}")
print("")
print(f"Tensor tns2 \n {tns2}")
print("")

resultado = torch.mm(tns1, tns2)
print(resultado.size())
print("")
print(resultado)
```

```python
""""
crie um tensor aleatório tns1 com a dimensionalidade 7 x 7 x 3
e um outro tensor aleatório tns2 de 147 x 1.
Modificando apenas tns1 some os dois tensores.
"""
import torch
tns11 = torch.randn(7,7,3)
tns22 = torch.randn(147, 1)
print(f"Tensor tns11 \n {tns11}")
print(f"Tensor tns22 \n {tns22}")

tns11 = tns11.view(-1, 1)
print(f"Tensor tns11 modificado \n {tns11}")
soma = tns11 + tns22
print(f"Soma: {soma}")
```

```python
import torch

# 1. Criação dos tensores com as dimensões solicitadas
tns11 = torch.randn(7, 7, 3)
tns22 = torch.randn(147, 1)

print(f"Dimensão original tns11: {tns11.shape}")
print(f"Dimensão original tns22: {tns22.shape}\n")

# 2. Modifica tns11 para o formato (147, 1) para alinhar com tns22
tns11_modificado = tns11.view(-1, 1)
# OBS.: view(-1, 1): O -1 diz ao PyTorch para calcular automaticamente o tamanho daquela dimensão.
# Como 7 × 7 × 3 = 147, o formato vira (147, 1).

# 3. Soma os dois tensores
soma_achatada = tns11_modificado + tns22

# 4. Restaura para o formato original (7, 7, 3)
resultado_final = soma_achatada.view(7, 7, 3)

print(f"Dimensão após a soma e restauração: {resultado_final.shape}")
print(f"\nResultado Final:\n{resultado_final}")
```
